```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Generate synthetic data for visualization
x = np.linspace(0, 10, 100)
y = np.sin(x)
noise = np.random.normal(0, 0.1, size=x.shape)
y_noisy = y + noise
```

```python
# Line plot
plt.figure(figsize=(6, 4))
plt.plot(x, y, label='True Function', linestyle='--')
plt.scatter(x, y_noisy, color='red', alpha=0.5, label='Noisy Data')
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.title("Line Plot with Noisy Data")
plt.legend()
plt.show()
```

```python
# Histogram
random_data = np.random.randn(1000)
plt.figure(figsize=(6, 4))
sns.histplot(random_data, bins=30, kde=True, color='blue')
plt.title("Histogram with KDE")
plt.xlabel("Value")
plt.ylabel("Frequency")
plt.show()
```

```python
# Scatter plot
x_scatter = np.random.rand(100)
y_scatter = np.random.rand(100)
plt.figure(figsize=(6, 4))
plt.scatter(x_scatter, y_scatter, c=y_scatter, cmap='viridis', alpha=0.7)
plt.colorbar(label='Color Intensity')
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.title("Scatter Plot with Color Mapping")
plt.show()
```

```python
# Box plot
data_box = [np.random.rand(50) * i for i in range(1, 6)]
plt.figure(figsize=(6, 4))
sns.boxplot(data=data_box)
plt.title("Box Plot of Random Data")
plt.xlabel("Category")
plt.ylabel("Values")
plt.show()
```
