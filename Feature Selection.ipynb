### Why value error might happen?
1. Mismatch in dimensions: X, y have proper/compatible dimensions. X- 2D(sample i.e rows and features) & y = 1D (sample)
2. Improper Datatype: X and y both should have numerical data type, if not -> error i.e value error
3. Incorrect target Dimensions/Shape: Instead of (sample) you have (sample, 1) -2D -> How to flatten 2D to 1D? -> by .flatten()
4. Invalid number of the features selected (value for k) < total number of features in dataset

```python
# Apply SelectKBest for chi-square test
from sklearn.feature_extraction.text import CountVectorizer # New library
from sklearn.feature_selection import SelectKBest, chi2
```

```python
X_data = [
    "My name is Sandhya.",
    "I am very smart",
    "Always into creativity",
    "I am ambivert"
]
y = [1,0,1,1]

vectorizer = CountVectorizer()
X = vectorizer.fit_transform(X_data)

select_k_best = SelectKBest(score_func = chi2, k=2)
X_new = select_k_best.fit_transform(X, y)

selected_features = select_k_best.get_support(indices = True)
feature_names = vectorizer.get_feature_names_out()
selected_features_names = feature_names[selected_features]
```
